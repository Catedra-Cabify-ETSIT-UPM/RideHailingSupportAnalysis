```python
import sys
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

#pd.set_option("display.max_colwidth", -1)
df = pd.read_csv('../../data/uber_support.csv')
df.shape
```

```python
df.drop_duplicates(inplace=True, subset="tweet")
df.shape
df.columns
```

```python
df['dateDay'] = pd.to_datetime(df['date']).dt.date
df
```

```python
df_p = df.groupby(df['dateDay']).size().reset_index(name='Count')
df_p
```

## Daily tweets

```python
figure(num=None, figsize=(80, 6), dpi=80, facecolor='w', edgecolor='k')
df_plot = df_p.set_index('dateDay')['Count'].plot.bar()
df_plot.set_xlabel('Date of tweets')
df_plot.set_ylabel('Number of tweets')
df_plot.set_title('Number of Custumer Tweets per day')
```

## Month tweets

```python
df_p.dateDay = pd.to_datetime(df_p.dateDay)
dm = df_p.groupby(pd.Grouper(key='dateDay', freq='1M')).sum() # groupby each 1 month
dm.index = dm.index.strftime('%B')
dm
```

```python
import numpy as np
import pandas as pd

figure(num=None, figsize=(40, 6), dpi=80, facecolor='w', edgecolor='k')

# Year
months = dm['Count'].keys().tolist()
# Name of the Quantities corresponding to the observations(Y Axis)
tweets = dm['Count'].tolist()


df = pd.DataFrame({
    'tweets': tweets

}, index=pd.date_range(start='2020/01/01', end='2021/01/01',
                       freq='M'))
ax = df.plot.area(legend = None, linewidth=2)
plt.grid()
ax.set_xlabel("Months 2020")
ax.set_ylabel("Number of tweets")
```
