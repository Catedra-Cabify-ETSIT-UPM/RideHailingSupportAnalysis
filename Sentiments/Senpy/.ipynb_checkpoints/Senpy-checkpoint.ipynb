```python
import os
from urllib.parse import urlunsplit
import requests
from IPython.display import Code
import json

import pandas as pd
import joblib
from time import time
```

# Load Dataset

```python
df_uber = pd.read_csv("../../Data_Extract&Analysis/uber_support_csv_with_duplicates/uber_support.csv")
df_uber.drop_duplicates(inplace=True, subset="tweet")
tweets = df_uber['tweet'].tolist()
df_uber
```

# Extract Sentiment

```python
polarities_sentiment140 = []
i = 0
start_time = time()

for tweet in tweets:
    i = i + 1
    if(i % 1000 == 0):
        print(i)

    url = urlunsplit(['http', 'localhost:5000', '/api', '', ''])
    params = {
    'algo': 'sentiment140',
    'input': tweet,
    'language': "en",
     }
    response = requests.get(url, params=params).json()
    polarity = response['entries']
    polarities_sentiment140.append(polarity)
    
print(len(polarities_sentiment140))
print("-------Time used for calculating sentiment with Sentiment140: {} s --------".format(time()-start_time))
joblib.dump(polarities_sentiment140, "polaritiessent140.pkl")
```

# Analize Sentiment140
