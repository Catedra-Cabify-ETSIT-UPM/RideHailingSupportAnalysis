```python
import joblib
from gensim.corpora import Dictionary
import gensim.corpora as corpora
from gensim.corpora import MmCorpus
from gensim import corpora, models, matutils
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans, KMeans, Birch, DBSCAN

import pandas as pd


from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import os
from sklearn import metrics


from time import time



from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.pipeline import Pipeline

from sklearn import metrics
```

```python
corpus = MmCorpus("../LDA/LDA//LDA6/results/variables/corpus.corp")
texts = joblib.load('../LDA/LDA/LDA6/results/variables/texts.pkl')
fulltexts = joblib.load('../LDA/LDA/LDA6/fulltexts.pkl')
model_list_6 = joblib.load('../LDA/LDA/LDA6/results/ldamodel/model_list_6.pkl')
lda_model = model_list_6[6]
dictionary = Dictionary.load("../LDA/LDA/LDA6/results/variables/dictionary.dict")
```

```python
document_prob = [lda_model[doc][0] for doc in corpus]
#document_prob
```

```python
document_prob_new = []
index = [x for x in range(0,7)]

for tweet in document_prob:
    index_tweet = []
    tweet_prob_new = []
    for tupla in tweet:
        index_tweet.append(tupla[0])
        tweet_prob_new.append(tupla)
        
    zero_values = np.array(np.setdiff1d(index, index_tweet))
    for i in zero_values:
        tweet_prob_new.append((i, 0.0))
    
    document_prob_new.append(tweet_prob_new)

for tweet in document_prob_new:
    tweet.sort(key=lambda x:x[0])

matrix = []
for tweet in document_prob_new:
    tweet_prob = []
    for value in tweet:
        tweet_prob.append(value[1])
    matrix.append(tweet_prob)

# Document - Matrix with 7 dimensions (probability assignament to each topic)  
matrix_np = np.array(matrix)
```

```python
#from scipy import sparse

#matrix = sparse.csr_matrix(matrix_np)
```

```python
# Sabemos que K optimo = 7; Vamos a optimizarlo variando n iters,
# cambiando las dimensiones de PCA (cuidado con la varianza)
# Veremos las m√©tricas shillouete, davis...
```

```python
# Modelos que podemos optimizar debido al dataset: Kmeans, MiniBatchKmeans,
# y Birch
```

```python
preprocessor = Pipeline(
         [
        ("pca", PCA(n_components=2, random_state=0)),
         ]
)
```

```python
clusterer = Pipeline(
            [
                ("kmeans",
                    KMeans(
                        n_clusters=7,
                        random_state=0,
                        n_init = 10,
                        ),
                ),
                
                ("miniBatch",
                    MiniBatchKMeans(
                        n_clusters=7,
                        random_state=0,
                        n_init = 10,
                    ),
                ),
                
                ("birch",
                    Birch(
                          n_clusters=None, 
                          branching_factor = 50, 
                          threshold = 0.5,
                    ),
                ),
                
            ]
 )
```

```python
pipe = Pipeline(
        [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
        ]
     )
```

```python
path = "./results/tunning"
try:
    if not os.path.isdir(path):
        os.makedirs(path)
except OSError:
    print ("Creation of the directory %s failed" % path)
else:
    print ("Successfully created the directory %s " % path)
```

```python
# Kmeans
```

```python
n_init_range = np.array([1, 5, 10, 15, 20, 50, 75, 100])


silhouette_scores = []
calinski_harabasz_scores = []
davies_bouldin_scores = []
inertia_scores = []

df_res_kmeans = pd.DataFrame(columns = ['Dimensions' , 'n_init', 
                                'Silhouette' ,'Inertia',
                                'Calinski_harabasz', 
                                'Davies_bouldin']
                      )

length = 0

for n in range(2, 8):
    
    range_array = []
    range_array.append(n)
    
    for i in n_init_range:
        
        temporal_array = []
        temporal_array.append(n)
        
        
        pipe["preprocessor"]["pca"].n_components = n
        pipe.fit(matrix_np)
        pipe["clusterer"]["kmeans"].n_init = i
        pipe.fit(matrix_np)
        
        inertia_coef = pipe["clusterer"]["kmeans"].inertia_

        silhouette_coef = silhouette_score(
            pipe["preprocessor"].transform(matrix_np),
            pipe["clusterer"]["kmeans"].labels_,
        )
        
        calinski_harabasz_coef = metrics.calinski_harabasz_score(
            pipe["preprocessor"].transform(matrix_np),
            pipe["clusterer"]["kmeans"].labels_,
        )
        
        davies_bouldin_score = metrics.davies_bouldin_score(
            pipe["preprocessor"].transform(matrix_np),
            pipe["clusterer"]["kmeans"].labels_,
        )

        # Add metrics to their lists
        temporal_array.append(i)
        temporal_array.append(silhouette_coef)
        temporal_array.append(inertia_coef)
        temporal_array.append(calinski_harabasz_coef)
        temporal_array.append(davies_bouldin_score)
        
        # Tengo una fila de datos
       # print(temporal_array)
        
        df_res_kmeans.loc[length] = temporal_array 

        length = length + 1
           
        
```

```python
df_res_kmeans.to_csv('./results/tunning/df_res_kmeans.csv', index=False)
```
